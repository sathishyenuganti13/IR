```python
from collections import defaultdict

def finding(text):
    words = []
    curr_word = ""
    for ch in text:
        if ch.isalnum() or ch == '_':
            curr_word += ch
        else:
            if curr_word:
                words.append(curr_word.lower())
            curr_word = ""
    if curr_word:
        words.append(curr_word.lower())
    return words
```

```python
def create(documents):
    pos_index = defaultdict(lambda: defaultdict(list))

    for doc_id, doc in enumerate(documents):
        terms = finding(doc)
        for position, term in enumerate(terms):
            pos_index[term][doc_id].append(position)

    return pos_index
```

```python
def print_index(pos_index):
    for term, postings in pos_index.items():
        print(f"Term: {term}")
        for doc_id, positions in postings.items():
            print(f"  Doc ID: {doc_id}, Positions: {positions}")
```
